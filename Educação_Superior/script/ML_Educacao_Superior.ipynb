# Análise de dados - Microdados do ENEM 2021

O Dataset utilizado na análise pode ser encontrado no site do INEP

```python
# Importando as bibliotecas a ser utilizadas

import pandas as pd
```

```python
EndIn = 'Z:/Desenvolvimento/Machine Learning - Educação Superior/microdados_enem_2021/DADOS/MICRODADOS_ENEM_2021.csv'
EndOut = 'Z:/Desenvolvimento/Machine Learning - Educação Superior/resultado/D_Enem_2021.csv'
```

O conjunto de Microdados do ENEM 2021 é um conjunto grande de informação, contendo aproximadamente 1.5 GB de tamanho. 
Para importa-lo inteiramente para o projeto no Jypter Notebook, necessitaria de muito mais memória que meu dispositivo possuía. 
Logo, para identificar quais colunas eu precisaria para fazer a análise que pretendia, eu utilizei o parâmetro 
nrows = 10 do comando pd.read_csv. 

```python
#Importar apenas as 10 primeiras linhas do dataset, 

D_Enem_2021 = pd.read_csv(EndIn, sep=";",encoding="ISO-8859-1", nrows=10)
display(D_Enem_2021)
```

```python
# Mostrar o nome das colunas do dataset, assim eu posso analisar quais colunas iria precisar
print(D_Enem_2021.columns)
```

```python
# Foram escolhidas as colunas do dataset que preciso analisar

Colunas = ['NU_INSCRICAO', 'NU_ANO', 'CO_MUNICIPIO_ESC', 'NO_MUNICIPIO_ESC', 'CO_UF_PROVA',
           'CO_MUNICIPIO_PROVA','NO_MUNICIPIO_PROVA', 'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 
           'NU_NOTA_MT', 'NU_NOTA_REDACAO']
```

```python
#Importar apenas colunas de interesse do dataset, 

openData = pd.read_csv(EndIn, sep=";",encoding="ISO-8859-1", usecols = Colunas,chunksize = 10 ** 6)
D_Enem_2021 = pd.DataFrame()
for row in openData:
    D_Enem_2021 = pd.concat([row,D_Enem_2021],ignore_index=True)
        
display(D_Enem_2021)
```

```python
# Verificar o uso de memória do computador ao importar o Dataset
D_Enem_2021.info(memory_usage='deep')
```

```python
# Filtrando dados so para Goiás
D_Enem_2021 = D_Enem_2021[D_Enem_2021['CO_UF_PROVA']==52]
D_Enem_2021.reset_index(inplace=True)
del D_Enem_2021['index']
display(D_Enem_2021)
```

```python
# Quantos municipios diferetes tiveram provas 
D_Enem_2021.groupby('NO_MUNICIPIO_PROVA')['NO_MUNICIPIO_PROVA'].count()
```

```python
# Quantas pessoas fizeram o enem em 2021 por município
DfMunProva = D_Enem_2021['NO_MUNICIPIO_PROVA']
DfMunProvaTotal = DfMunProva.value_counts()
DfMunProvaTotal
```
